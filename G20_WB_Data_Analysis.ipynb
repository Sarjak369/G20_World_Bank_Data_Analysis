### Outlier Proportion¶

We can check outliers with the help of boxplot

```python
fig, axs = plt.subplots(nrows=4, ncols=4, figsize=(14,10))
axs = axs.flatten()

for i, col in enumerate(selected_cols):
    sns.boxplot(data=dff, x=col, ax=axs[i])
    axs[i].set_title(col)

plt.tight_layout()
plt.show()
```

Obervations:

We can see that almost every column except for 'Mobile phone subscriptions', 'Imports of goods and services', 'Exports of goods and services',  there are outliers present in all the other columns..

Some columns have high outliers like 'GDP_BN' and 'GNI_BN' which is expected because this is something economic indicator and some countries are growing at much rapid pace and they're having high GDP compared to the other countries..

Let's take 'Access to electricity',
These are G20 countries right, so it is expected that access to electricity should be at higher level. But, within that we see quite disparity and we can see the outliers even on the lower side..

So, these are the outliers which we can see with the help of boxplots.

But, we don't know the proportion of outliers.
If we want to understand the proportion of outlier and have to numerically capture that, we can probably define a function and can calculate whether each record is an outlier or not an outlier and we can calculate the number of rows out of total rows and consider it as in the proportion of outliers in the data set or maybe in a particular column.


```python
def outlier(x):
    q1=x.quantile(0.25)
    q3=x.quantile(0.75)
    iqr = q3 - q1
    return ( x < q1-1.5*iqr) | (x>q3+1.5*iqr)
```

```python
outlier_percentages = []

for col in selected_cols:
    x = dff[col]
    outliers = outlier(x)
    outlier_percentage = np.mean(outliers)*100
    outlier_percentages.append(outlier_percentage)
    
    
result_df = pd.DataFrame({'Column': selected_cols, 'Outlier Percentage': outlier_percentages})
result_df = result_df.sort_values('Outlier Percentage', ascending=False)
```

```python
result_df.round()
```

We can see that the outlier percentage in 'Access to electricity' column is the highest and then the 'Agricultural land area' column and so on and so forth.

We can visually see the outlier percentages in terms of a bar graph

```python
plt.figure(figsize=(11,5))
sns.barplot(x=result_df['Column'], y=result_df['Outlier Percentage'])
plt.title("Outlier Percentage")
plt.xticks(rotation=90)
plt.show()
```

### Top 3 countries

```python
# We have a data from 2015 to 2020..and we have 16 different columns..
# So, for each of the parameter, how countries have been ranked every year
# Let's see the top 3 countries..

def top_3(data, parameter):
    year_list = data['year'].dt.year.sort_values().unique()
    result_df = pd.DataFrame(columns=year_list,index=[1,2,3])
    
    for year in year_list:
        df_year = data[data['year'].dt.year == year]
        value = df_year.groupby('country')[parameter].sum().sort_values(ascending=False).head(3).index
        result_df[year] = value
    
    return result_df
```

### Life Expectancy

Let's understand the life expentancy.
For that, we need to create our own data using pivot table.. 
So, we can use pd.pivot_table from pandas library..

Pivot table is a great method to visualize and to analyze your data. It will pivot around one, two or multiple columns and it will give you values across other columns by grouping to that particular pivot call..

So here we are going to use pivot table. 

```python
life_expectancy = pd.pivot_table(columns='year_o', index='country', values='Life expectancy', data=dff, aggfunc='mean').round(1)

```

```python
life_expectancy.columns
```

```python
life_expectancy
```

```python
life_expectancy['perc_chg'] = round((life_expectancy[2020] - life_expectancy[2015])*100/life_expectancy[2015],1)

```

```python
life_expectancy = life_expectancy.sort_values(ascending=False, by='perc_chg')

```

```python
life_expectancy
```

```python
plt.figure(figsize=(11,5))
sns.barplot(x=life_expectancy.index, y=life_expectancy['perc_chg'], palette='viridis')
plt.title("% Change in Life Expectancy from 2015 to 2020")
plt.xticks(rotation =90)
plt.show()
```
